```python
lst = [1, 2, 3, 4, 5, 6] # Создаёт список lst с числами от 1 до 6.

result = []
for el in lst: # Перебирает каждый элемент el из lst
  if el%2 == 0:  # Проверяет, делится ли число на 2 без остатка
    result.append(el)  # Если условие выполняется, добавляет элемент в result

print(result)
```

```python
lst = [1, 2, 3, 4, 5, 6]

# 1. filter() применяет лямбда‑функцию к каждому элементу lst
# 2. Лямбда проверяет, чётное ли число (остаток от деления на 2 равен 0)
# 3. filter() оставляет только элементы, где условие True
# 4. list() преобразует результат filter() в список
result = list(filter(lambda x: x % 2 == 0, lst))

print(result)  # Вывод: [2, 4, 6]
```

```python
 # обычный деклоративный стиль
lst = [1, 2, 3, 4, 5, 6] # создаем список

result = [] # результат записываем сюда
for el in lst:  # for элемент in последовательность: цикл, который выполняется для каждого элемента
  result.append(el**2)  # добавляем квадрат каждого элемента в result

print(result)
```

```python
list(filter(lambda x: x%2 == 0, lst))   # Исходный список: [1, 2, 3, 4, 5, 6].Это коллекция данных, которую мы будем фильтровать.
                                        # lambda — ключевое слово для создания анонимной (без имени) функции.

# x — параметр функции (каждый элемент из списка lst).
# x % 2 — операция взятия остатка от деления на 2.
# == 0 — проверка, равен ли остаток нулю (т. е. число чётное).
# Вся лямбда‑функция возвращает True для чётных чисел и False для нечётных.
```

```python
#isinstance Функция isinstance() проверяет, является ли объект экземпляром
#указанного класса или типа данных (либо подклассом этого класса). Она возвращает:
#True, если объект соответствует классу/типу;
#False — в противном случае.
#lst = [1.5, '1e-5', 1.23e+1, 5, '1.108']
# eval(x) выполняет строку, переданную ей в качестве аргумента, как выражение на языке Python, и возвращает результат выполнения.
# Функция анализирует строку, проверяет её синтаксис, преобразует в исполняемый код
# и выполняет его в текущем контексте (если не заданы ограничения)

print(list(map(lambda x: x if isinstance(x, (int, float)) else eval(x), lst)))
```

```python
# Функция reduce(function, iterable[, initializer])
# последовательно применяет функцию к элементам итерабельного объекта, «сворачивая» его в одно значение.
from functools import reduce # reduce- уменьшить
```

```python
[1, 2, 3, 4, 5]
```

```python
lst = [1, 10, -1, 5, 6, 3]
```

```python
# сумма всего ряда, работает как аккумулятор
reduce(lambda x, y: x+y, lst)
```

```python
# возвр. x если x>y иначе y,находим самое больш.число в списке
reduce(lambda x, y: x if x > y else y, lst)
```

```python
class Person:
  age = 0
  name = ''

Mike = Person()
Mike.name = 'Mike'
Mike.age = 15

Joe = Person()
Joe.name = 'Joe'
Joe.age = 20

Mary = Person()
Mary.name = 'Mary'
Mary.age = 13

lst = [Mike, Joe, Mary]
```

```python
lst
```

```python
# сортировка по ключу .age
lst.sort(key=lambda x: x.age)
```

```python
for person in lst:
  print(person.name)
```

```python
lst = [['Andron', 5], ['Ivan', 7], ['Kirill', 3], ['Aaron', 1]]
```

```python
sorted(lst) # сортирует в алфавитном порядке
```

```python
sorted(lst, key=lambda x: x[1])
# Аргумент key используется в функциях sorted() и
# методе list.sort() для указания функции, которая будет применена к
# каждому элементу перед сравнением. Эта функция задаёт критерий сортировки — то,
# по чему именно нужно сортировать элементы.
```

```python
d = {   #словарь активности магнитного поля
    1: {
        'Фон': 10,
        'Пик': 20,
        'Подъем': -3,
        'Спад': 5
    },
    2: {
        'Фон': 10,
        'Пик': 20,
        'Подъем': 300,
        'Спад': 5
    },
    3: {
        'Фон': 10,
        'Пик': 20,
        'Подъем': 300,
        'Спад': 555
    }
}
```

```python
d1 = {
        'Фон': 10,
        'Пик': 20,
        'Подъем': -3,
        'Спад': 5
    }
```

```python
#воспольз ф.max кот.
max(d1, key=d1.get)
# Метод get() используется для безопасного получения значения из словаря по ключу.
#Если ключ существует, возвращается соответствующее значение;
# если ключа нет — возвращается None или указанное по умолчанию значение (не вызывает исключение).
# key- что явл. критерием отбора
# вывести ключ кот. соотв. максиммальн.знач.
#
```

```python
list(map(lambda x: max(x, key=x.get), d.values()))
# map-применить для каждого элемента словаря из d
# d.values()- для значений словаря
```
